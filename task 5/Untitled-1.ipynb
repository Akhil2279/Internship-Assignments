```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("sales_data.csv")

# Show first few rows
print(" Sales Data Preview:")
print(df.head())

# Total Sales by Category
category_sales = df.groupby("Category")["Sales"].sum()
print("\n Total Sales by Category:")
print(category_sales)

# Total Quantity Sold by Product
product_quantity = df.groupby("Product")["Quantity"].sum()
print("\n Quantity Sold by Product:")
print(product_quantity)

# Daily Sales Trend
daily_sales = df.groupby("Date")["Sales"].sum()
print("\n Daily Sales Trend:")
print(daily_sales)

# --- Plot Charts ---
# Sales by Category
category_sales.plot(kind="bar", title="Total Sales by Category")
plt.ylabel("Sales")
plt.show()

# Quantity by Product
product_quantity.plot(kind="bar", title="Quantity Sold by Product")
plt.ylabel("Quantity")
plt.show()

# Daily Sales Trend
daily_sales.plot(kind="line", marker="o", title="Daily Sales Trend")
plt.ylabel("Sales")
plt.show()
```
